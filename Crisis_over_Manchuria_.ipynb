##The Manchurian Incident and the League
-------------
**18 September 1931 bomb explosion on the South Manchurian Railway** (Mukden incident)


> Leads to fighting and the occupation of the city and bombing of Jinzhou

- Japans approach: Bilateral talks with China. Argument: Japan took place to guard and protect the rights and interests of Japanese citizens. **No territorial designs** 
- China's approach: as early as the 21. Sep brought a case to the attention of the League

> 1931 fight starts between Chinese and Japanese troops in Shangai > Bombing of Shangai > leads to shifting of minds of most observers who were sympathetic to Japan's exasperation in Manchuria.

**While the league sent a commission the Guandong Army had "engineered the creation of the later on puppet state Manchukuo.** 

Japan's argument: Japanese recognition of and assistance to Manchukue would bring development to the region and supports stability for the region. 


- The commission's final report: upheld Japanese colonial interest in the region, but the military operations of the Japanese army cannot be regarded as legitimate self-defense. 
- Independence was **only** made possible because of the military presence of the Japanese army. 

Most Japanese who interacted with the League directly hoped for a solution were Japan could achieve its imperial goals but at the same time remain part of the League. 

- 42 to 1 (Japan) vote to accept the report of the Committee
> "Symbolic Japanese break from the world order"


**The League from the start had been build, especially in regard to the unique Sino-Japanese relationship. (...) superficial observation, empty speculation, and mistaken interpretations of international law and treaties.**

- Following heavy lobbying by various Japanese former League members all over Europe and America.  














##The Reaction of Internationalists 
------------
Some members of the Japanese association spoke out for a further commitment to the League. For exp. Yamakawa Tadao and Yoshina Sakuzo.

Argument: 
- If Japan stays in the League it will have more stable relations with the US and can avoid being internationally isolated. 
- Increased military expenses will bring pressure on the Japanese economy 
- Efforts made in the League over the last years would be a big backlash for Japans international position

**Again lobbying at international labor, business and community groups at campuses churches, and civic clubs.**





##Nitobe Inazo 
------------------
"the League as a universal organ, to deal with all disputes arising throughout the world (...) there should not be too many competing organs:"

- public and international apology for Japanese continental policy
- proposal of representation of Wilsons doctrine of natural self-determination
- explain but not justify Japans behaviour, correct misinformation, and confront US prejudices with regard to Japan-China politics. 

Argument: China can not function as a sovereign state in the modern sense.

- "We have learned many things from America, especially in dealing with neighbouring unstable governments."



##Ishii Kikajiro
----------
- persuade the world of Japans rightful policy 
- claimed military actions in Manchuria were carried out without prior knowledge of the chief of state or the Foreign Ministry
- ergo. the situation in China should not be viewed as a sign of Japanese militarism or Japanese disaffection for the League.
- doubt of competence of the "tools" of the League to asses the situation in the appropriate way

Argument: China has historicall made outlandish claims over border regions and treated the regions with disinterest and neglect.
